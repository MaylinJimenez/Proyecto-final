```python
mpl.rcParams["axes.spines.bottom"]=True #spines son los ejes
mpl.rcParams["axes.spines.top"]= False
mpl.rcParams["axes.spines.left"]= True
mpl.rcParams["axes.spines.right"]= False
```

```python
#-------------------------------------------------------------------------------------------------#
#Parametros: dfParam -> Es el set de datos completo,
#            nomColumna -> El nombre de la variable por la cual voy a agrupar los datos.
#            orden -> True para que ordenado de menor a mayor, False para se ordene de mayor a menor
#-------------------------------------------------------------------------------------------------#
    
def agrupaYcuenta(dfParam, nomColumna, orden): #gf es un df para graficar
    gf=dfParam.groupby([nomColumna]).count()
    gf=gf.iloc[:,:1] #permite filtrar y seleccionar columnas y observaciones, al colocarlo asi [:] él entiende que se quieren todas las observaciones
    gf.rename({"serial_number":"cantidad"}, axis=1, inplace=True)
    gf=gf.sort_values('cantidad',ascending=orden)
    return gf
```

<h4>Pregunta de Negocio #4</h4>
<h5  font-family: Times New Roman"> ¿Cual es el costo promedio de las propiedades residenciales y comerciales?</h5>

```python
# Filtrar datos y calcular promedio
residential_avg = dfpropiedades[dfpropiedades['property_type'] == 'RESIDENTIAL']['assessed_value'].mean()
commercial_avg = dfpropiedades[dfpropiedades['property_type'] == 'COMMERCIAL']['assessed_value'].mean()

# Ajustar valores mínimos para evitar barras invisibles
commercial_avg = commercial_avg if commercial_avg > 0 else 1

# Crear DataFrame con los valores a graficar
data = {'Tipo de Propiedad': ['RESIDENTIAL', 'COMMERCIAL'],
        'Costo Promedio': [residential_avg, commercial_avg]}

# Graficar
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)
ax.bar(data['Tipo de Propiedad'], data['Costo Promedio'], color=['#a2caa5', '#f4a460'])

# Configuración del eje Y para que siempre se vean las barras
ax.set_ylim(0, max(residential_avg, commercial_avg) * 1.1)

# Configuración del gráfico
ax.set_xlabel("Tipo de Propiedad", fontsize=12, fontweight="bold")
ax.set_ylabel("Costo Promedio", fontsize=12, fontweight="bold")
ax.set_title("Gráfico 4: Costo Promedio entre Propiedades Residenciales y Comerciales", fontsize=14, fontweight="bold")

# Añadir texto en las barras
for i, valor in enumerate([residential_avg, commercial_avg]):
    ax.text(i, valor + 1, f'{valor:.2f}', ha='center', va='bottom', fontsize=10)

# Mostrar gráfico
plt.show()
```

<h4>Pregunta de Negocio #5</h4>
<h5>¿Cuál es el tipo de propiedades residenciales que se vende con mayor frecuencia? Donde y con que costo promedio?</h5>

```python
# Filtrar valores nulos antes de contar y calcular el costo promedio
dfpropiedades_filtrado = dfpropiedades.dropna(subset=['residential_type', 'assessed_value'])

# Agrupar por tipo de residencia y calcular la cantidad de ventas y el costo promedio
residential_counts = dfpropiedades_filtrado['residential_type'].value_counts()
residential_avg_cost = dfpropiedades_filtrado.groupby('residential_type')['assessed_value'].mean()

# Crear figura y eje
fig, ax = plt.subplots(figsize=(15,6.9), dpi=100)

# Definir lista de colores específica (ajusta según la cantidad de categorías)
colores = ["#b2cbae","#76877d","#88b8a9","#b2cbae","#dbddb4"] 

# Gráfico de barras con colores personalizados
ax.bar(residential_counts.index, residential_counts.values, color=colores[:len(residential_counts.index)], alpha=0.8)

# Etiquetas y títulos
titulo = "Gráfico 5: Tipo de propiedad residencial más vendida y su costo promedio"
plt.suptitle(titulo,
             horizontalalignment='left',
             x=0.001,
             y=0.87,
             transform=fig.transFigure,
             fontsize=20,
             fontweight="bold",
             fontfamily='serif')

subTitulo = "Cantidad de ventas por tipo de residencia y su costo promedio"
ax.set_title(subTitulo,
             ha='left',
             x=0.001,
             y=0.80,
             transform=fig.transFigure,
             fontsize=12,
             fontweight="regular",
             fontfamily="serif")

ax.set_xlabel("Tipo de Residencia",
              ha='left',
              fontsize=10,
              fontweight="regular",
              fontfamily="serif")

ax.set_ylabel("Cantidad de Ventas",
              ha='left',
              fontsize=10,
              fontweight="regular",
              fontfamily="serif")

# Añadir etiquetas de costo promedio
for i, v in enumerate(residential_counts.index):
    ax.text(i, residential_counts[v], f"{residential_avg_cost[v]:.2f}", ha='center', fontsize=10, fontweight="bold")

plt.xticks(rotation=45)
plt.subplots_adjust(top=0.75)
fig.savefig("Grafico_residential_type_mas_vendido", dpi=300, bbox_inches='tight')
plt.show()
```
